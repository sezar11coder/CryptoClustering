```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
numerical_columns = df_market_data.select_dtypes(include=['float64', 'int64']).columns
df_numerical = df_market_data[numerical_columns] 
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numerical)
```

```python
# Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(df_scaled, columns=numerical_columns, index=df_market_data.index)



# Copy the crypto names from the original DataFrame
if 'coin_id' in df_market_data.columns:
    crypto_names = df_market_data[['coin_id']].copy()
else:
    crypto_names = df_market_data.index.to_frame(name="coin_id")  # If coin_id is the index


# Set the coin_id column as index
if 'coin_id' in df_market_data.columns:
    df_market_data.set_index('coin_id', inplace=True)

# Also update the scaled DataFrame to have the same index
df_scaled.index = df_market_data.index


# Display the scaled DataFrame
df_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    
# 2. Fit the model to the data using `df_market_data_scaled`
    kmeans.fit(df_scaled)
# 3. Append the model.inertia_ to the inertia list
    inertia.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k': k_values, 'Inertia': inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.plot(df_elbow['k'], df_elbow['Inertia'], marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve for Optimal k')
plt.xticks(k_values)
plt.grid(True)

# Show the plot
plt.show()
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Best value is k=4 From the graph, I look for the sharpest drop followed by a more gradual decrease

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
best_k = 4
kmeans_model = KMeans(n_clusters=best_k, random_state=42, n_init=10)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans_model.fit(df_scaled)
```

```python

# Print the resulting array of cluster values
cluster_values = kmeans_model.predict(df_scaled)
print(cluster_values)
```

```python
# Create a copy of the scaled DataFrame
df_scaled_copy = df_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_scaled_copy['Predicted_Cluster'] =cluster_values 

# Display the copy of the scaled DataFrame
df_scaled_copy.head()
```

```python
import hvplot.pandas
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
scatter_plot = df_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Predicted_Cluster",  # Color points based on cluster labels
    hover_cols=["Cluster"],  # Add crypto name for identification
    width=800,
    height=500,
    title="Crypto Clustering Based on 24h and 7d Price Changes"
)

# Display the scatter plot
scatter_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_transformed = pca.fit_transform(df_scaled.iloc[:, :-2])
df_pca = pd.DataFrame(pca_transformed, columns=['PCA1', 'PCA2', 'PCA3'], index=df_scaled.index)

# View the scaled PCA data
df_pca.head()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

df_explained_variance = pd.DataFrame(
    explained_variance, 
    index=['PCA1', 'PCA2', 'PCA3'], 
    columns=['Explained Variance']
)

df_explained_variance
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** This shows how much of the original data's variance is captured by each PCA component. Typically, the first few components hold the most information

```python
# Create a new DataFrame with the PCA data.
df_pca_final = df_pca.copy()

# Copy the crypto names from the original scaled DataFrame
df_pca_final['coin_id'] = df_scaled.index

# Set the coin_id column as index
df_pca_final.set_index('coin_id', inplace=True)

# Display the scaled PCA DataFrame
df_pca_final
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values_pca = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of 
for k in k_values_pca:
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans_pca = KMeans(n_clusters=k, random_state=42, n_init=10)
# 2. Fit the model to the data using `df_market_data_pca`.
    kmeans_pca.fit(df_pca_final)
# 3. Append the model.inertia_ to the inertia list
    inertia_pca.append(kmeans_pca.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {'k': k_values_pca, 'Inertia': inertia_pca}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data_pca)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
plt.figure(figsize=(8, 5))
plt.plot(df_elbow_pca['k'], df_elbow_pca['Inertia'], marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve for Optimal k (PCA Data)')
plt.xticks(k_values_pca)
plt.grid(True)

# Show the plot
plt.show()
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** < k=4 > After k = 4, the decrease in inertia slows down, indicating diminishing returns from adding more clusters.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, the best k-value using PCA-transformed data (k = 4) is the same as the one found using the original scaled data (k = 4)

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
kmeans_pca_model = KMeans(n_clusters=4, random_state=42, n_init=10)
```

```python
# Fit the K-Means model using the PCA data
kmeans_pca_model.fit(df_pca_final)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
predicted_clusters_pca = kmeans_pca_model.predict(df_pca_final)


# Print the resulting array of cluster values.
print(predicted_clusters_pca)
```

```python
# Create a copy of the scaled PCA DataFrame
df_pca_final_predictions = df_pca_final.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
df_pca_final_predictions["PredictedCluster"] = predicted_clusters_pca

# Display the copy of the scaled PCA DataFrame
df_pca_final_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
df_pca_final_predictions.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="PredictedCluster",
    title = "Scatter Plot by Crypto Segment - PCA=2"
)
```
